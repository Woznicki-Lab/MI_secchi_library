# Tidy data

```python

import os
from google.colab import files

import pandas as pd
import altair as alt
import numpy as np
import io
from google.colab import data_table
import datetime
from pandas.tseries.offsets import Day

data_table.enable_dataframe_formatter()
```

```python
#upload file
uploaded = files.upload()
df = pd.read_csv(io.BytesIO(uploaded['MAIN_all_lakes_zstats_S2_jrc50_19-21_1.csv']))

#drop unneeded cols
df = df.iloc[0:, 2:]
df = df.drop(['AreaSqKm', 'COUNTY_Nam', 'COUNTY_Zon', 'ComID', 'EDU_Name', 'EDU_ZoneID', 'Elevation', 'FCode','FDate','FType','GNIS_ID','GNIS_Name','HU12','HU12_Name','HU12_ZoneI','HU4','HU4_Name','HU4_ZoneID','HU6','HU6_Name',
              'HU6_ZoneID','HU8','HU8_Name','HU8_ZoneID','LakeConnec','Lake_Area_','Permanent_','ReachCode','STATE','STATE_Zone','State_Name'],axis=1)

#wide to long
test = df.melt(['lagoslakei'], var_name = "imageID", value_name = "values")
#drop nas
test = test.dropna()

#split imageID into col variables
test[["satellite", "date", "delete_date", "tile", "band"]] = test["imageID"].str.split('_', expand=True)
test = test.drop(['delete_date', 'imageID'], axis=1)
test[["capture_date", "delete"]] = test["date"].str.split('T', expand=True)
test = test.drop(['date', 'delete'], axis=1)
#format date
test["capture_date"] = pd.to_datetime(test["capture_date"], format='%Y%m%d')

#remove duplicates
test = test.drop_duplicates(keep='first')

#long to wide format
test = test.pivot_table(values='values',
                        index=['lagoslakei', 'satellite', 'tile', 'capture_date'],
                        columns=['band'],
                        aggfunc='mean').reset_index()

#display
data_table.DataTable(test, num_rows_per_page=10)
```
